```python
import pandas as pd
import matplotlib.pyplot as plt



```

```python
df = pd.read_parquet("../data/agg/binance/BTC-USDT/BTC-USDT_1m_2022-2025.parquet")
df['dt'] = pd.to_datetime(df['timestamp'], unit ='ms')

df['ret'] = df["close"]
```

```python
df = df.set_index('dt')
```

```python
df['ret'] = df['close'].pct_change().fillna(0.0)
df
```

```python
import numpy as np
df['logret'] = np.log(df["close"]).diff().fillna(0.0)
df
```

```python
df["cumret"]=(1+df["ret"]).cumprod()-1
df
```

```python
df['cumlogret']=df['logret'].cumsum()
df
```

```python
df["cumlogret_exp"] =np.exp(df["cumlogret"])-1
df
```

```python
df["cumret"].plot(figsize=(15,6));
df[["cumret", "cumlogret_exp"]].plot(figsize=(15,6))
```

```python
df["cumret_change"] = df["cumret"].diff()
df["cumret_change"].nlargest(5)
df["cumret_change"].nsmallest(5)
```

```python
running_max = df["cumret"].cummax()
df["dd"] = df["cumret"] / running_max - 1
df["dd"].plot(figsize=(15,6));
```
